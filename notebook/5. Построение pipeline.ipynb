```python
import pandas as pd
import numpy as np
from collections import Counter
import matplotlib.pyplot as plt
```

```python
from tslearn.clustering import TimeSeriesKMeans
from scipy.spatial.distance import euclidean
from fastdtw import fastdtw
import time
from datetime import timedelta
from tqdm import tqdm
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score, adjusted_rand_score, adjusted_mutual_info_score
```

После того как мы восстановили временной ряд, требуется проводить модификацию

```python
df_general = pd.read_csv('../data/2. Восстановленные')
dates = pd.to_datetime(df_general.columns,format='%d.%m.%Y')
scores = {}
```

```python
def modify_time_series(time_series,k = 2, m=0.5, mode=None):
    mean = np.mean(time_series)
    std_dev = np.std(time_series)

    lower_bound_2sigma = mean - k * std_dev
    upper_bound_2sigma = mean + k * std_dev
    lower_bound_1sigma = mean - m*std_dev
    upper_bound_1sigma = mean + m*std_dev

    modified_time_series = np.copy(time_series)

    modified_time_series[(modified_time_series >= lower_bound_1sigma) & (modified_time_series <= upper_bound_1sigma)] = mean
    
    indices = []
    for index, value in enumerate(modified_time_series):
        if value == mean:
            indices.append(index)
    
    return modified_time_series, np.array(indices)


def count_elements_in_list_of_arrays(arrays):
    flattened_list = [item for sublist in arrays for item in sublist]
    counter = Counter(flattened_list)
    return counter
```

```python
def calculate_dtw_distance(ts1, ts2):
    distance, _ = fastdtw(ts1, ts2)
    return distance

def initialize_clusters(data, k):
    np.random.seed(0)
    centroids_indices = np.random.choice(len(data), k, replace=False)
    return [data[i] for i in centroids_indices]

def assign_to_clusters(data, clusters):
    assignments = []
    for ts in data:
        min_distance = float('inf')
        assigned_cluster = None
        for centroid_idx, centroid in enumerate(clusters):
            distance = calculate_dtw_distance(ts, centroid)
            if distance < min_distance:
                min_distance = distance
                assigned_cluster = centroid_idx
        assignments.append(assigned_cluster)
    return assignments

def update_centroids(data, assignments, k):
    new_clusters = []
    for cluster_idx in range(k):
        cluster_data = [data[i] for i, assignment in enumerate(assignments) if assignment == cluster_idx]
        new_centroid = np.mean(cluster_data, axis=0)
        new_clusters.append(new_centroid)
    return new_clusters

def k_means_with_dtw(data, k, max_iterations):
    clusters = initialize_clusters(data, k)
    for _ in tqdm(range(max_iterations)):
        old_clusters = clusters
        assignments = assign_to_clusters(data, clusters)
        clusters = update_centroids(data, assignments, k)
        if np.array_equal(old_clusters, clusters):
            break
    return assignments, clusters

def evaluate_clustering(data, assignments, centroids):
    silhouette = silhouette_score(data, assignments)
    davies_bouldin = davies_bouldin_score(data, assignments)
    calinski_harabasz = calinski_harabasz_score(data, assignments)
    return silhouette, davies_bouldin, calinski_harabasz
```

```python
def get_modify(df, min_level_sigma):

    # Модификация
    arr_indexes_mean = []
    for index, row in df.iterrows():
            modified_row, indices = modify_time_series(row, m = min_level_sigma)
            arr_indexes_mean.append(indices)
            df.loc[index] = modified_row
    return arr_indexes_mean, df


def get_counter_index_df(arr_indexes_mean):
    arrays = arr_indexes_mean
    counter = count_elements_in_list_of_arrays(arrays)
    sum_counter = sum(counter.values())
    counter_df = pd.DataFrame(sorted([[k,v/sum_counter] for k,v in counter.items()],key=lambda x: x[0]),columns=['индекс', 'значение'])
    counter_df['date'] = dates
    counter_df['значение'].sum()
    counter_df['%'] = counter_df['значение']*100
    counter_df['количество'] = counter_df['значение'] * sum_counter
    counter_df = counter_df.sort_values(by='значение', ascending=False)
    counter_df['аккамуляция %'] = counter_df['%'].cumsum()
    return counter_df
```

```python
min_level_sigma = 0.17
df = df_general.copy()
arr_indexes_mean, df = get_modify(df, min_level_sigma)
counter_df = get_counter_index_df(arr_indexes_mean)
```

```python
score_row = []
scores[min_level_sigma] = []
for p in tqdm([2,4,6,8,10]):
    # Пример использования
    select_df = counter_df[counter_df['аккамуляция %']<p] # ЗДЕСЬ МЫ ДОЛЖНЫ ВЫБИРАТЬ
    indexes_for_cut = counter_df.loc[counter_df['аккамуляция %']<p]['индекс'].values
    selected_columns = df.drop(df.columns[indexes_for_cut], axis=1)
    
    data = selected_columns.values[:50]
    k = 3
    max_iterations = 10

    start_time = time.time()
    assignments, clusters = k_means_with_dtw(data, k, max_iterations)
    end_time = time.time()
    elapsed_time = end_time - start_time
    formatted_time = str(timedelta(seconds=elapsed_time))
    print(f"Time taken: {formatted_time}")
    silhouette, davies_bouldin, calinski_harabasz = evaluate_clustering(data, assignments, clusters)
    score_row = [silhouette, davies_bouldin, calinski_harabasz ]
    scores[min_level_sigma].append(score_row)
```
